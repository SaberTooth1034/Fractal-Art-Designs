```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def func1(x,y):
  return (0.,0.16*y)
def func2(x,y):
  return (0.85*x + 0.04*y, -0.04*x + 0.85*y + 1.6)
def func3(x,y):
  return (0.2*x - 0.26*y, 0.23*x + 0.22*y + 1.6)
def func4(x,y):
  return (-0.15*x + 0.28*y, 0.26*x + 0.24*y + 0.44)


  
funcs=[func1, func2, func3, func4]
pnts=100000

x=0
y=0
xlist=[]  
ylist=[]

for i in range(pnts):
    #probability related to each function
  func=np.random.choice(funcs,p=[0.01, 0.85, 0.07, 0.07])
  x,y=func(x,y)
  xlist.append(x)
  ylist.append(y)

plt.scatter(xlist, ylist, color='green', s=0.2)
```

```python
b, l=300,300
fern=np.zeros((b,l)) #empty array in which we put the image
x,y=0,0
for i in range(pnts): #over 100,000 points
    func=np.random.choice(funcs,p=[0.01, 0.85, 0.07, 0.07])
    x,y=func(x,y)
    idx, idy=int(b/2 + x*b / 10), int(y * l / 12)#recentering the points
    fern[idy,idx]=1
plt.imshow(fern[::-1,:], cmap=cm.Greens) #printing in reverse order otherwise it will be upside down
plt.show()
```

```python
#mandelbrot set

def function1 (c,z):
    return z**5 + c
```

```python
c1=complex(1,2)
z0=0
function1(c1,z0)
z1=function1(c1,z0)
```

```python
def diverge(c, z=0, n=20, B=20000):
    c=complex(*c)
    for i in range(n):
        z=function1(c,z)
        if abs(z)>B:
            return 1
    return 0
```

```python
x_min, x_max= -2,1
y_min, y_max= -1,1
res=200

xx, yy=np.meshgrid(np.linspace(x_min, x_max,res), np.linspace(y_min, y_max, res))
np.set_printoptions(2)
points=np.c_[xx.ravel(),yy.ravel()]
```

```python
mandelbrot=[diverge(c) for c in points]
mandelbrot=np.array(mandelbrot)
mandelbrot=mandelbrot.reshape(res,res)
plt.contourf(mandelbrot)
```

```python
def MB(k):
    def function1(c, z, k):
        return z**k+c
    
    def diverge(c, z=0, n=20, B=20000):
        c=complex(*c)
        for i in range(n):
            z=function1(c, z, k)
            if abs(z)>B:
                return 1
        return 0
    
    x_min, x_max= -4,2
    y_min, y_max= -2,2
    res=200

    xx, yy=np.meshgrid(np.linspace(x_min, x_max,res), np.linspace(y_min, y_max, res))
    np.set_printoptions(2)
    points=np.c_[xx.ravel(),yy.ravel()]

    mandelbrot=[diverge(c) for c in points]
    mandelbrot=np.array(mandelbrot)
    mandelbrot=mandelbrot.reshape(res,res)
    plt.contourf(mandelbrot)
    plt.savefig(f'mandelbrot_{int(k*1000)}.png')
```

```python
from turtle import *

shape('turtle')
speed(0)
left(90)
def tree(size, levels, angle):
    if levels==0: #base case
        color("green")
        dot(size)
        color('black')
        return
    forward(size)
    
    right(angle)
    tree(size*0.8, levels-1, angle*1.05)
    
    left(angle*2)
    tree(size*0.8, levels-1, angle*1.05)
    
    right(angle) # now we are in the middle
    backward(size) #to trace back to starting point

tree(90, 6, 30)

mainloop()
```

```python
shape('turtle')
speed(0)
#left(90)

def snowflake(sides, length):
    for _ in range(sides):
        snowflakeside(length, sides)
        right(360/sides)

def snowflakeside(length, levels):
    if levels==0:
        forward(length)
        return
    
    length/=3.0
    
    snowflakeside(length, levels-1)
    left(60)
    
    snowflakeside(length, levels-1)
    right(120)

    
    snowflakeside(length, levels-1)
    left(60)

    
    snowflakeside(length, levels-1)
    
snowflake(3,200)

mainloop()
    
    
    

    
    
```
